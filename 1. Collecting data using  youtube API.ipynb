```python
#step 5
  #lets create a funtion to call/collect 2nd APIs for parameters likes, views etc.

def get_video_details(video_id):

#step 4
    URL_videos_remaining_parameters =  "https://www.googleapis.com/youtube/v3/videos?id="+video_id+"&part=statistics&key="+API_key  
    reponse_videos_remaining_parameters = requests.get(URL_videos_remaining_parameters).json()

    view_count = reponse_videos_remaining_parameters['items'][0]['statistics']['viewCount']
    like_count = reponse_videos_remaining_parameters['items'][0]['statistics']['likeCount']

    return view_count,like_count

```

```python
#step 7 to call APIs

def videos(df):      #step3(after calling the 2nd API)
  #calling out API for parameters like videoid, title and upload date.

    #step3
  pageToken = "" #parameter allows to navigate through pages.

    #step2
  URl = "https://www.googleapis.com/youtube/v3/search?key="+API_key+"&channelId="+channel_ID+"&part=snippet,id&order=date&maxResults=10000"+pageToken 
  #https://developers.google.com/youtube/v3/docs/search/list
  #need to provide URL to access the data and authenticate through that API.
  #note : every API has its own URL parameters, so need to read documentions.

    #step1
  response = requests.get(URl).json()  #this will get the data from API in json file format.
  time.sleep(1)

#step 3
  # We need to collect parameters of each videos of SDSU channel, lets use for loop
  for video in response["items"]:
    if video["id"]["kind"]=='youtube#video':
      video_id = video["id"]["videoId"]
      video_title = video["snippet"]["title"]
      video_title = str(video_title).replace("&amp:","")
      video_upload_date = video["snippet"]["publishedAt"]
      video_upload_date = str(video_upload_date).split("T")[0]

#step 6
    #making 2nd API call through function
      like_count, view_count = get_video_details(video_id)
    
      print(video_id)
      print(video_title)
      print(video_upload_date)
      print(view_count)
      print(like_count)
      #save data in data frame df
      df = df.append({ "video_id": video_id, "video_title": video_title, "video_upload_date":video_upload_date,
      "view_count":view_count, "like_count":like_count},ignore_index= True    
      )
  return df

```

```python
#step 8
#create pandas data frame
df = pd.DataFrame(columns = ["video_id","video_title","video_upload_date","view_count","like_count"])
df = videos(df)
```
